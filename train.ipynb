```python
from sklearn.model_selection import train_test_split
import pandas as pd

data = pd.read_csv("./data/augmented_feature_train.csv", index_col=0)
data.head()
```

```python
import xgboost as xgb
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import pandas as pd


X = data.drop(columns=["Label"])
y = data["Label"]

# Split the data into training and validation sets (75% training, 25% validation)
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.25, random_state=42)

X_train.shape, X_val.shape, y_train.shape, y_val.shape
```

```python
params = {
    'objective': 'multi:softmax',
    'num_class': 3,  # Three classes as you mentioned
    'booster': 'gbtree',
    'eval_metric': 'mlogloss',
    'eta': 0.01,
    'max_depth': 10,
    'subsample': 0.8,
    'colsample_bytree': 0.8,
    'seed': 42
}

dtrain = xgb.DMatrix(X_train, label=y_train)
dval = xgb.DMatrix(X_val, label=y_val)
evals = [(dtrain, 'train'), (dval, 'eval')]
num_rounds = 100000

bst = xgb.train(params, dtrain, num_rounds, evals, early_stopping_rounds=50, verbose_eval=True)
y_val_pred = bst.predict(dval, iteration_range=(0, bst.best_iteration+1))
```

```python
accuracy = accuracy_score(y_val, y_val_pred)
print(f"Validation Accuracy: {accuracy * 100:.2f}%")
```

```python
test_data['LOCATION'] = None

# Iterate through rows of the dataframe
for idx, row in test_data.iterrows():
    lat, lon = row['LAT'], row['LON']
    
    # Check each key in the location dictionary
    for location, coords in location_per_latlon.items():
        if (lat, lon) in coords:
            test_data.at[idx, 'LOCATION'] = location
            break

test_data.shape[0]
```

```python
test_data['YEAR'] = test_data['time'].apply(lambda x: int(str(x)[:4]))
test_data['MONTH'] = test_data['time'].apply(lambda x: int(str(x)[4:6]))
test_data['DAY'] = test_data['time'].apply(lambda x: int(str(x)[6:]))
test_data.shape[0]
```

```python
test_data["PRECT"] = test_data["PRECT"] * 3600 * 1000
test_data.shape[0]
```

```python
test_data = test_data.drop(columns=["LAT", "LON", "time"])
test_data.shape[0]
```

```python
test_data["SOUTHERN_HEMISPHERE"] = test_data["LOCATION"] <= 2
test_data.shape[0]
```

```python
import numpy as np

radians = (test_data['MONTH'] - 1) * (2 * np.pi / 12)

# Create sine and cosine columns
test_data['SIN_MONTH'] = np.sin(radians)
test_data['COS_MONTH'] = np.cos(radians)

test_data = test_data.drop(columns=["MONTH"])
test_data.shape[0]
```

```python
# Magnitude
test_data["WIND850_MAGNITUDE"] = np.sqrt(test_data["U850"]**2 + test_data["V850"]**2)
test_data["WINDBOT_MAGNITUDE"] = np.sqrt(test_data["UBOT"]**2 + test_data["VBOT"]**2)

# Wind direction
radians_850 = np.arctan2(test_data["V850"], test_data["U850"])
test_data["WIND850_SIN"] = np.sin(radians_850)
test_data["WIND850_COS"] = np.cos(radians_850)

radians_BOT = np.arctan2(test_data["VBOT"], test_data["UBOT"])
test_data["WINDBOT_SIN"] = np.sin(radians_BOT)
test_data["WINDBOT_COS"] = np.cos(radians_BOT)

test_data = test_data.drop(columns=["U850", "V850", "UBOT", "VBOT"])
test_data.shape[0]
```
